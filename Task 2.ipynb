# Dividing data into test and train

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

# Training a model

```python
model = make_pipeline(
    StandardScaler(),
    SVC(kernel='rbf')
)
model.fit(X_train,y_train)
```

```python
y_pred_training = model.predict(X_test)
print(confusion_matrix(y_test,y_pred_training))
print(accuracy_score(y_test,y_pred_training))
```

```python
print("Classification report")
print(classification_report(y_test, y_pred_training))
```

```python
result = pd.DataFrame({'Actual': y_test, 'Predicted':y_pred_training})
result = pd.concat([result,X_test],axis= 1)
display(result)
```
